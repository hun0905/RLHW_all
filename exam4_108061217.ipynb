```python
def chooseAction(state, qtable, epsilon):
  if rd.rand()>=epsilon:
    return np.argmax(qtable[state[0],state[1]])
  else:
    return rd.randint(4)
```

```python
class ShortCorridor:
    """
    Short corridor environment
    """
    def __init__(self):
        self.reset()

    def reset(self):
        self.state = 0

    def step(self, go_right):
        o1=True
        o2=0
        ###### Your code here ######
        if self.state in [1,2]:
          go_right=1-go_right  ##reverse
        if go_right:
          self.state+=1
        else:
          self.state-=1
        self.state=np.clip(self.state,0,3)
        if self.state==3:
          o1=True
          o2=0
        else:
          o1=False
          o2=-1
        ################################
        return [o1, o2]
```

### 2.4(15%)Actor-Critic
Please apply Actor-Critic in the environment with $\gamma=1 $,  learning rate for policy $\alpha_\theta=0.001$, learning rate for value state $\alpha_v=0.01$ over 500 episodes.
 The output 'returns' is a list consisting of return 'G' from each episode.



```python
GAMMA = 1.0
ALPHA_theta = 0.001
ALPHA_v = 0.01

# Actor critic
def chooseAction_pg(pi):
    ###### Your code here ######
    action=np.random.choice(2, p=pi)
    ################################
    return action

# Actor critic
def gradient_ln( action, policy):
  ###### Your code here ######
  tmp = np.zeros(2)
  tmp[action]=1
  grad = tmp - policy
  ################################
  return grad

# Actor critic
def AC():
    ###### Your code here ######
    h_table = np.zeros([2, 2])
    v_table = np.zeros([2])
    pi = np.ones([2,2]) * 0.5
    returns = []
    env = ShortCorridor()
    for episode in range(500):
      state = 0
      G=0
      env.reset()
      is_terminal=False
      while not is_terminal:
        action = chooseAction_pg(pi[state,:])
        state_n,reward=env.step(action)
        is_terminal = state_n
        delta = reward + GAMMA*v_table[state_n]-v_table[state]
        h_table[state,:] = h_table[state,:] + ALPHA_theta*delta*gradient_ln(action,pi[state,:])
        v_table[state] = v_table[state] + ALPHA_v*delta
        pi[state,:] = np.exp(h_table[state,:])/sum(np.exp(h_table[state,:]))
        G += reward
        state = state_n
      returns.append(G)
    ################################
    return returns
```
